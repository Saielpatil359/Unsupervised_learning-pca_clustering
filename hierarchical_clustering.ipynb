```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
df = pd.read_csv('student_data.csv')
df.head()
```

```python
X = df[["Attendance","AssignmentScore","ExamScore"]]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
plt.figure(figsize=(10,6))
dendrogram = sch.dendrogram(sch.linkage(X_scaled,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Students')
plt.ylabel('Distance')
plt.show()
```

```python
hc = AgglomerativeClustering(
    n_clusters=3,
    linkage='ward'
)
labels = hc.fit_predict(X_scaled)
df['Cluster']=labels
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_scaled[:,0],X_scaled[:,2],c=labels)
plt.xlabel('Attendance(scaled)')
plt.ylabel('Exam Score(scaled)')
plt.title('Studnet Clusters')
plt.show()
```

```python
df.groupby('Cluster')[['Attendance', 'AssignmentScore', 'ExamScore']].mean()
```
